```python
# Define a function to convert phone numbers
def format_phone_number_method1(input_number):
    
    if(input_number):
        pattern = r'\(?(\d{3})\)?[-.\s]*(\d{3})[-.\s]*(\d{4})$'
        
        if(re.match(pattern,input_number)):
            matches = re.search(pattern,input_number)
            # Extract the different parts of the phone number
            area_code = matches.group(1)
            first_part = matches.group(2)
            second_part = matches.group(3)

            # Concatenate the parts into the desired format
            formatted_phone_number = f"({area_code}) {first_part}-{second_part}"

            return formatted_phone_number
        else:
            return 'not matched'
    else:
        return ''
```
